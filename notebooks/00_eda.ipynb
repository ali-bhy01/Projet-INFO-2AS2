## 1 — Loading & schema validation

The CSV has no header. Columns are: `date ; time ; open ; high ; low ; close ; volume`  
Timezone: CET (already local — the 09:15 candle is directly readable, no conversion needed).

```python
df = pd.read_csv(
    DATA_PATH,
    sep=";",
    header=None,
    names=["date", "time", "open", "high", "low", "close", "volume"],
)

df["datetime"] = pd.to_datetime(df["date"] + " " + df["time"], format="%d/%m/%Y %H:%M")
df = df.set_index("datetime").drop(columns=["date", "time"])
df = df.sort_index()

print(f"Rows       : {len(df):,}")
print(f"Date range : {df.index.min().date()}  →  {df.index.max().date()}")
print(f"Columns    : {list(df.columns)}")
print(f"NaN count  : {df.isnull().sum().sum()}")
df.head(3)
```
